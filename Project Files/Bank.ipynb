```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Bank_Churn.csv')
df
```

```python
df = df.drop(columns = ['CustomerId', 'Surname',])
df = df.drop_duplicates()
df = df.dropna()
df
```

```python
df = df.rename(columns = {'CreditScore' : 'Credit Score', 'Geography' : 'Country',
                          'NumOfProducts' : 'Number Of Products', 'HasCrCard' : 'Credit Card',
                          'IsActiveMember' : 'Active Member', 'EstimatedSalary' : 'Salary'})
df
```

```python
df['Credit Card'] = df['Credit Card'].replace({1 : True,
                                               0 : False})
df['Active Member'] = df['Active Member'].replace({1 : True,
                                               0 : False})
df['Exited'] = df['Exited'].replace({1 : True,
                                    0 : False})
df
```

```python
df = df.to_excel('Cleaned_Bank.xlsx', index = False)
```
